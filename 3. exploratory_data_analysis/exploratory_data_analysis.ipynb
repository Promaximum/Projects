Есть квартиры, с числом конмнат "0" (всего 197). Такого быть не может. Изучим число квадратных метров в данных комнатах

```python
data.query('rooms == 0')['total_area'].describe()
```

Количество квадратных метров в таких квартирах почти 30, значит, скорее всего, это однокомнатные квартиры

```python
for i in range(1,5): #Напишем цикл для определения того, сколько в квартире примерно комнат от 1 до 4 и выведем для каждой описывающие параметры
    print(f'{i} -комнатная квартира:', data.query('rooms == @i')['total_area'].describe())
```

Мы определили интервалы для количества комнат в квартире: 
* 1-комнатная (менее 40 квадратных метров)
* 2-комнатная (40-60 квадратных метров)
* 3-комнатная (60-80 квадратных метров)
* 4-комнатная (более 80 квадратных метров)

```python
#напишем функцию, по которой мы категоризируем квартиры по числу комнат
data.loc[(data['rooms']==0) & (data['total_area']<40), 'rooms'] = 1
data.loc[(data['rooms']==0) & (data['total_area']>40) & (data['total_area']<60), 'rooms'] = 2
data.loc[(data['rooms']==0) & (data['total_area']>60) & (data['total_area']<80), 'rooms'] = 3
data.loc[(data['rooms']==0) & (data['total_area']>60) & (data['total_area']<130), 'rooms'] = 4

```

Удалим квартиру, где 371 квадратных метра, так как мы не знаем, сколько там комнат

```python
print('Число строк до удаления', data.shape)
data.drop(data.loc[(data['rooms'] ==0) & (data['total_area']==371)].index, inplace=True, axis=0)
print('Число строк после удаления', data.shape)
```

```python
data.reset_index(inplace=True, drop=True)
print(data.info())
```

Видно, что в некоторых строках вместо "ё" стоит "е". Заменим все "ё" на "е" при помощи библиотеки re и цикла for

```python
import re
i = 0 #счётчик букды Ё и ё
for index_str in data.loc[(data['locality_name'].isna()==False), 'locality_name'].index:
    loc_str = data.loc[index_str, 'locality_name']
    if ('ё' or 'Ё') in loc_str:
        i+=1
        loc_str = re.sub('ё', 'е', loc_str)
        loc_str = re.sub('Ё', 'Е', loc_str)
    data.loc[index_str, 'locality_name'] = loc_str
print()
print('Количество строк, где была произведена замена:', i)
```
